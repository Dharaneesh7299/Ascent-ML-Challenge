```python
# 2. Load Model
import sys
import os
sys.path.append('src')
from inference import Upscaler
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

MODEL_PATH = 'netG_epoch_25.pth'

if os.path.exists(MODEL_PATH):
    upscaler = Upscaler(model_path=MODEL_PATH)
    print("‚úÖ Model Loaded Successfully")
else:
    print(f"‚ùå Model file {MODEL_PATH} not found. Please upload it to Colab.")
    # Fallback for testing without model
    upscaler = Upscaler(model_path=None)
```

```python
# 3. Upload Test Image
from google.colab import files
uploaded = files.upload()
filename = list(uploaded.keys())[0]
print(f"Processing {filename}...")
```

```python
# 4. Run Inference & Vizualize
lr, sr = upscaler.predict(filename)

# Display
plt.figure(figsize=(15, 8))

plt.subplot(1, 2, 1)
plt.title('Original (Low Res / Input)')
plt.imshow(lr)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Super Resolved (AI Output)')
plt.imshow(sr)
plt.axis('off')

plt.tight_layout()
plt.show()

# Save output
sr.save('output_sr.png')
print("Saved output to output_sr.png")
```
